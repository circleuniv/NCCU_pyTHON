## `scikit-learn` 真實世界數據

`scikit-learn` 內建一些真實世界的數據, 可以讓你玩玩, 他們稱做 "Toy Datasets"。有哪些可以參考 [scikit-learn 官網](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)的說明。

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
````

````python
from sklearn.datasets import load_diabetes
````

## 讀入 diabetes 數據

````python
diabetes=load_diabetes()
````

## 用 feature_names 看到數據中所有的 features。發現這數據有10個features!

````python
diabetes.feature_names
````

### 根據說明 Y 是糖尿病進展的定量估計值

````python
X=diabetes.data
Y=diabetes.target
````

````python
X
````

````python
len(X)
````

````python
x_train, x_test, y_train, y_test = train_test_split( X, Y, 
                                                    test_size=0.3, 
                                                    random_state=87)
````

````python
regr=LinearRegression()
````

````python
regr.fit(x_train,y_train)
````

````python
y_predict=regr.predict(x_test)
````

````python
plt.scatter(y_test,y_predict)
plt.plot([50,300],[50,300],'r')
plt.xlabel("True measure")
plt.ylabel("Predicted measure")
````

## 畫出不同的 feature 對 糖尿病估計值 的影響

````python
plt.figure(figsize=(8,10)) #設定畫布大小
for i, feature in enumerate(diabetes.feature_names):
    plt.subplot(5, 2, i+1)
    plt.scatter(X[:,i],Y,s=1)
    plt.ylabel("measure")
    plt.xlabel(feature)
    plt.tight_layout() 
````
