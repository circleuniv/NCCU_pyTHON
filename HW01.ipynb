* 給定x 範圍
* 畫出 $\frac{sinx}{2}$ 在 $x=[-\pi,\pi]$ 的圖形

## 給定x 範圍, 並印出x

```python
x=np.linspace(-np.pi,np.pi)
```

```python
x
```

## 依所給定x算出 y=$\frac{sinx}{2}$, 並印出y

```python
y=np.sin(x)/2
```

```python
y
```

## 畫美美的圖

```python
plt.plot(x,y);
```
