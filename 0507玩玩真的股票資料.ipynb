```python
df=web.DataReader('AAPL','stooq', start='2022-2-17', end='2023-2-17')
df.head()
```

# 計算報酬率
## $$\frac{P_{t} - P_{t-1}}{P_{t-1}}$$

```python
P.diff()/P #跟前一期的差別. 要算 報酬率 再 除 原來的P
```

```python
P[-100:].plot()
```

# 移動平均

```python
P.rolling(window=20).mean() # 二十天的移動平均
```

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
