# 準備模擬的資料

### 做一條直線
### 比如說
$$f(x)=1.2x+0.8$$

### 準備好50個點

```python
x=np.linspace(0,5,50)
```

### 畫出圖形來。

```python
y=1.2*x+0.8
```

```python
plt.scatter(x,y)  #只是要把點畫出來, 我們觀測到的資料
plt.plot(x,y,'r') #真實世界的完美函數
```

## 加入 noise 項, 看起來更真實

### 大概的想法就是，我們真實世界的問題，化成函數，我們假設背後有個美好的函數。但相信我們很少看到真實世界的資料這麼漂亮。在統計上，我們就是假設
### $$f(x)+e(x)$$
### 也就是都有個noise 項。

```python
y=1.2*x+0.8 + 0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x, 1.2*x+0.8,'r')
```

# 載入線性迴歸工具

```python
from sklearn.linear_model import LinearRegression
```

## 開 線性迴歸機 的方法

```python
regr=LinearRegression()
```

```python
X=x.reshape(50,1) #改變array形狀的方法
```

```python
X
```

```python
regr.fit(X,y)
```

## 開始預測的方法

```python
Y=regr.predict(X)
```

```python
plt.scatter(x,y) #真實世界觀測到的樣子
plt.plot(x,Y,'r') #模擬的樣子
plt.plot(x, 1.2*x+0.8,'g') #完美世界的函數
```
