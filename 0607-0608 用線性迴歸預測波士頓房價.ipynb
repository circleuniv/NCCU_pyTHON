## `scikit-learn` 真實世界數據

`scikit-learn` 內建一些真實世界的數據, 可以讓你玩玩, 他們稱做 "Toy Datasets"。有哪些可以參考 [scikit-learn 官網](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)的說明。

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
````

### 讀入 boston 房價數據

````python
from sklearn.datasets import load_boston
````

````python
boston=load_boston()
````

# 資料裡到底有什麼

### features
 你可以用 feature_names 看到數據中所有的 features。發現這數據有13個features!



````python
boston.feature_names
````

````python
X=boston.data
Y=boston.target
````

````python
len(X)
````

````python
x_train, x_test, y_train, y_test = train_test_split( X, Y, 
                                                    test_size=0.3, 
                                                    random_state=87)
````

````python
regr=LinearRegression()
````

````python
regr.fit(x_train,y_train)
````

````python
y_predict=regr.predict(x_test)
````

````python
plt.scatter(y_test,y_predict)
plt.plot([0,50],[0,50],'r')
plt.xlabel("True Price")
plt.ylabel("Predicted Price")
````

### [畫多個圖]

````python
x=np.linspace(-10,10,200)
````

````python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
````

````python
plt.subplot(2,2,1) #放2x2排法的第一張圖
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,x)

plt.subplot(2,2,4)
plt.plot(x,x**2)
````

### 炫炫的畫出個別參數和 target 關係

````python
plt.figure(figsize=(8,10)) #設定畫布大小
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:,i],Y,s=1)
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout()    
````
