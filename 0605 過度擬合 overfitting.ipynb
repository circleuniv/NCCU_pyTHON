```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
x=np.linspace(0,5,50)
y=1.2*x+0.8+0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x,1.2*x+0.8,'r')
```

```python
X=np.linspace(0,5,1000)

def my_fit(n):
    Y=4*np.sin(n*X)+4  #中心點取4(+4), 振幅取4(乘以4)
    plt.scatter(x,y) # 真實世界觀察到的
    plt.plot(X,Y,'r')
    plt.show() #互動必打
```

```python
my_fit(5)
```

```python
interact(my_fit,n=(1,250)) # n愈大, 愈碰得到每個觀察的點 (這函數不好 叫 overfitiing)
```

### Overfitting 意思是說，我今天在訓練資料裡，學得非常好，可是，拿到真實世界的時候，發現說完全不行，預測的亂七八糟。
### 這時候，就叫 Overfitting
